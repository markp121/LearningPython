# Creating Functions

```python
def fahr_to_celcius(temp_f):
    temp_c = (temp_f - 32) * (5/9)
    return temp_c
```

```python
temp = 10
print(temp, " degrees F is ", fahr_to_celcius(temp), "degrees C" )
```

```python
def celcius_to_kelvin(temp_c):
    temp_k = temp_c + 273.15
    return temp_k
```

```python
def fahr_to_kelvin(temp_f):
    temp_c = fahr_to_celcius(temp_f)
    temp_k = celcius_to_kelvin(temp_c)
    
    return temp_k
```

```python
temp = 20

print(temp, " degrees F is ", fahr_to_kelvin(temp), "degrees K")
```

```python
import numpy
import matplotlib.pyplot as plt
import glob
```

```python
def check_data_ok(filename):
    data = numpy.loadtxt(filename, delimiter=",")
    min_values = numpy.min(data,axis=0)
    sum_min = numpy.sum(min_values)
    
    if sum_min == 0:
        print("data looks dodgy")
    else:
        print("looks ok")
        
def visualise(filename):
    data = numpy.loadtxt(filename, delimiter=",")
    
    fig = plt.figure(figsize=(10.0, 3.0))
    
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)
    
    axes1.set_ylabel("average")
    axes1.plot (numpy.mean(data, axis=0))
    
    axes2.set_ylabel("max")
    axes2.plot (numpy.max(data, axis=0))
    
    axes3.set_ylabel("min")
    axes3.plot (numpy.min(data, axis=0))
    
    fig.tight_layout()
    plt.show()
```

```python
filenames = sorted(glob.glob("data/inflammation-*.csv"))

for filename in filenames:
    visualise(filename)
    check_data_ok(filename)
```

# Testing functions

```python
def offset_mean(data, target_mean_value):
    """
    This function will offset the mean.
    
    Paramters:
    data, target_mean_value
    """
    offset_data = (data - numpy.mean(data)) + target_mean_value
    return offset_data
```

```python
z = numpy.zeros((2,2))
offset_z = offset_mean(z, 1)
print(offset_z)
```

```python
data = numpy.loadtxt("data/inflammation-01.csv", delimiter=",")
print(offset_mean(data, 1))
```

```python
print("original min, mean, max are: ", numpy.min(data), numpy.mean(data), numpy.max(data))
offset_data = offset_mean(data, 0)
print("offset min, mean, max are: ", numpy.min(offset_data), numpy.mean(offset_data), numpy.max(offset_data))
```
